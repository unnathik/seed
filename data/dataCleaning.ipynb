```python
import pandas as pd
```

```python
file = "environmental-dataset.csv"
dataset = pd.read_csv(file)
```

```python
dataset = dataset[dataset['Country'] == 'UNITED STATES OF AMERICA']
```

```python
dataset.drop(columns='Year', inplace=True)
```

```python
dataset.drop(columns=['Environmental Intensity (Sales)', 'Environmental Intensity (Op Inc)'], inplace=True)
```

```python
dataset.drop(columns=['Working Capacity', 'Fish Production Capacity', 'Crop Production Capacity', 'Meat Production Capacity'], inplace=True)
```

```python
dataset
```

```python
dataset['% Imputed'] = dataset['% Imputed'].str.replace('%', '').astype(float) / 100
```

```python
dataset.rename(columns={'% Imputed': 'Environmental Index'}, inplace=True)
```

```python
dataset.drop(columns='Wood Production Capacity', inplace=True)
```

```python
dataset.dropna(inplace=True)
```

```python
dataset['Total Environmental Cost'] = dataset['Total Environmental Cost'].str.replace(',', '').astype(int)
dataset['Total Environmental Cost'] = dataset['Total Environmental Cost'] + abs(dataset['Total Environmental Cost'].min())
```

```python
dataset['Total Environmental Cost'] = dataset['Total Environmental Cost'] / (dataset['Total Environmental Cost'].max())
```

```python
dataset['Biodiversity'] = dataset['Biodiversity'].str.replace(',', '').astype(int)
dataset['Biodiversity'] = dataset['Biodiversity'] + abs(dataset['Biodiversity'].min())
dataset['Biodiversity'] = dataset['Biodiversity'] / (dataset['Biodiversity'].max())
```

```python
dataset.drop(columns='Water production capacity (Drinking water & Irrigation Water)', inplace=True)
```

```python
import numpy as np

columns = ['Abiotic Resources', 'SDG 1.5', 'SDG 2.1', 'SDG 2.2', 'SDG 2.3', 'SDG 2.4', 'SDG 3.3', 'SDG 3.4', 'SDG 3.9', 'SDG 6', 'SDG 12.2','SDG 14.2','SDG 14.3', 'SDG 15.1', 'SDG 15.2', 'SDG 15.5']

for c in columns:
    print(c)
    dataset[c] = dataset[c].str.replace(',', '').astype(int)
    dataset[c] = dataset[c] + abs(dataset[c].min())
    dataset[c] = dataset[c] / (dataset[c].max())
```

```python
dataset['SDG 14.1'] = dataset['SDG 14.1'].str.replace(',', '')
dataset['SDG 14.1'] = dataset['SDG 14.1'].replace(' -   ', np.nan)
```

```python
dataset.dropna(inplace=True)
```

```python
dataset['SDG 14.1'] = dataset['SDG 14.1'].astype(int)
dataset['SDG 14.1'] = dataset['SDG 14.1'] + abs(dataset['SDG 14.1'].min())
dataset['SDG 14.1'] = dataset['SDG 14.1'] / (dataset['SDG 14.1'].max())
```

```python
dataset['SDG 14.c'] = dataset['SDG 14.c'].str.replace(',', '')
dataset['SDG 14.c'] = dataset['SDG 14.c'].replace(' -   ', np.nan)
dataset.dropna(inplace=True)
dataset['SDG 14.c'] = dataset['SDG 14.c'].astype(int)
dataset['SDG 14.c'] = dataset['SDG 14.c'] + abs(dataset['SDG 14.c'].min())
dataset['SDG 14.c'] = dataset['SDG 14.c'] / (dataset['SDG 14.c'].max())
```

```python
dataset
```

```python
dataset.to_csv('clean-environmental-dataset.csv', index=False)
```
